```python
gni = pd.read_csv('GNI.csv', sep = ',')
gni['Country or Area'] = gni['Country or Area'].replace(['Bolivia (Plurinational State of)'], 'Bolivia')
gni['Country or Area'] = gni['Country or Area'].replace(['China (mainland)'], 'China')
gni['Country or Area'] = gni['Country or Area'].replace(["Côte d'Ivoire"], "Cote d'Ivoire")
gni['Country or Area'] = gni['Country or Area'].replace(['Democratic Republic of the Congo'], 'Congo, Dem. Rep.')
gni['Country or Area'] = gni['Country or Area'].replace(['Egypt'], 'Egypt, Arab Rep.')
gni['Country or Area'] = gni['Country or Area'].replace(['Gambia'], 'Gambia, The')
gni['Country or Area'] = gni['Country or Area'].replace(['Iran, Islamic Republic of'], 'Iran, Islamic Rep.')
gni['Country or Area'] = gni['Country or Area'].replace(['Kingdom of Eswatini'], 'Eswatini')
gni['Country or Area'] = gni['Country or Area'].replace(["Lao People's Democratic Republic"], 'Lao PDR')
gni['Country or Area'] = gni['Country or Area'].replace(['Micronesia (Federated States of)'], 'Micronesia, Fed. Sts.')
gni['Country or Area'] = gni['Country or Area'].replace(['Slovakia'], 'Slovak Republic')
gni['Country or Area'] = gni['Country or Area'].replace(['Türkiye'], 'Turkiye')
gni['Country or Area'] = gni['Country or Area'].replace(['Yemen'], 'Yemen, Rep.')
gni['Country or Area'] = gni['Country or Area'].replace(['United Kingdom of Great Britain and Northern Ireland'], 'United Kingdom')
gni['Country or Area'] = gni['Country or Area'].replace(['Venezuela (Bolivarian Republic of)'], 'Venezuela, RB')
gni['Country or Area'] = gni['Country or Area'].replace(['Viet Nam'], 'Vietnam')
gni
```
